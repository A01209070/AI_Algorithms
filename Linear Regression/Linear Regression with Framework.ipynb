```python
import pandas as pd
from sklearn import preprocessing, linear_model
import matplotlib as plt
from sklearn.metrics import mean_squared_error

# define the colums of the attributes
columns = ["co", "pt08.s1", "nmhc", "c6h6", "pt08.s2", "nox", "pt08.s3", "no2", "pt08.s4", "pt08.s5", "t", "rh", "ah"]

# load the csv file of the data set of AirQualityUCI
data_frame = pd.read_csv('AirQualityUCI.csv', names = columns)
data_frame_y = data_frame[["no2"]]

print(data_frame.head())
print(data_frame_y.head())

# scale the data using the min_max_scaler preprocessing
temp_data_frame = data_frame.values
min_max_scaler = preprocessing.MinMaxScaler()
temp_data_frame_scaled = min_max_scaler.fit_transform(temp_data_frame)
data_frame = pd.DataFrame(temp_data_frame_scaled)

print(data_frame.head())

# split the data set between predictors and predctions
data_frame_x = data_frame[[0, 2, 3, 5, 10, 11, 12]] # attributes
print(data_frame_x.head())
print(data_frame_y.head())

# split the data in training and test sets
data_frame_x_train = data_frame_x[:-1872]
data_frame_x_test = data_frame_x[-1872:]
data_frame_x_validation = data_frame_x_test[-10:]
data_frame_y_train = data_frame_y[:-1872]
data_frame_y_test = data_frame_y[-1872:]
data_frame_y_validation = data_frame_y_test[-10:]

 

# linear regression object
regression = linear_model.LinearRegression()

# train the model
regression.fit(data_frame_x_train, data_frame_y_train)

# make predictions
predictions = regression.predict(data_frame_x_test) 

# print the results
print('Coefficients: \n', regression.coef_)
print('Mean Squared Error: %.2f' % mean_squared_error(data_frame_y_test, predictions))

predictions = regression.predict(data_frame_x_validation)
for i in range(len(predictions)):
  print("Predicted: ", predictions[i], "\tExpected: ", data_frame_y_validation.values[i])

  


```
